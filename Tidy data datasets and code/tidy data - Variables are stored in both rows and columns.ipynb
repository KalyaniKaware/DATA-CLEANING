```python
import pandas as pd

weather = pd.read_csv(r"C:\Users\kalya\Desktop\Study material\DATA CLEANING\Tidy data datasets and code\weather-raw.csv")
weather
```

```python
weather = weather.melt(id_vars = ['id','year','month','element'],var_name = 'day', value_name = 'temp')
```

```python
weather.head(3)
```

```python
weather.month.unique()
weather.day.unique()
weather.year.unique()
weather.day = weather.day.str[1:]
weather.columns
```

```python
weather = weather.pivot_table(index = ['id', 'year', 'month', 'day'], columns = 'element', values = 'temp')
```

```python
weather
```

```python
weather = weather.reset_index().rename_axis(None, axis = 1)
```

```python
#weather['date'] = weather['year'].astype(str) +'/'+ weather['month'].astype(str).str.zfill(2) +'/'+ weather['day'].astype(str).str.zfill(2)

weather=(weather.assign(date = lambda x: x.year.astype("str") +"/"+ x.month.astype("str").str.zfill(2) +"/"+ x.day.astype("str").str.zfill(2))
            .drop(["year", "month","day"],axis=1))
```

```python
weather['date'] = pd.to_datetime(weather['date'], format = '%Y/%m/%d')
```

```python
weather = weather.filter(items = ['id','tmax','tmin','date']).groupby(by='date').aggregate("mean").reset_index()
```
