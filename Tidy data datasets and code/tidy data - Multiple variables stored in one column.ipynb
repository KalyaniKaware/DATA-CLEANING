Series.str can be used to access the values of the series as strings and apply several methods to it.

```python
import pandas as pd
df_tb = pd.read_csv(r"C:\Users\kalya\Desktop\practise\Tidy data datasets\tb.csv")
df_tb.tail()
```

```python
df_tb.columns
```

```python
df_tb = df_tb.melt(id_vars = ['iso2', 'year', 'new_sp'],var_name = 'demographics',value_name = 'cases')
```

```python
df_tb.head()
```

```python
df_tb = df_tb.assign(
                        age = lambda x: df_tb.demographics.str[8:].astype(str),
                        gender = lambda x: df_tb.demographics.str[7:8].astype(str)
                    ).drop('demographics', axis = 1)

#Series.str can be used to access the values of the series as strings and apply several methods to it.
```

```python
df_tb.age.str.len().unique()
```

```python
df_tb.gender.unique()
```

```python
df_tb.age.unique()
```

```python
df_tb.update(pd.DataFrame({'age':[x[:2] + '-' + x[2:] if len(x) == 4 else (x) for x in df_tb.age]}))
```

```python
df_tb.sample(5)
```

```python
df_tb = df_tb.replace(to_replace = ['f','m','u','04','514','014'], value = ['Female','Male','Unknown','0-4','5-14','0-14']).dropna()
```
