```python
import pandas as pd
relinc = pd.read_csv(r"C:\Users\kalya\Desktop\practise\Tidy data datasets\relinc.csv")
relinc.sample()
```

```python
# Applying melt (to a long format) - unpivot

df_unpivot = relinc.melt(id_vars = 'religion',var_name = 'income', value_name = 'frequency')
df_unpivot.head()
```

```python
# Applying pivot_table (to a wide format)

df_pivot = df_unpivot.pivot_table(index = 'religion', columns = 'income', values = 'frequency')
df_pivot.head()
```

```python
df_pivot = df_pivot.reset_index().rename_axis(None, axis = 1)
df_pivot.head()
```
