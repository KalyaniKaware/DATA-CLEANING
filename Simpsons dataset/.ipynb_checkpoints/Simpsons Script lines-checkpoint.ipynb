```python
def f(text):
    return text.split(':')[1].split(')')[0]
```

```python
def g(text):
    if len(text.split('(')) >1:
        return text.split('(')[1]
```

```python
expression = script_lines['raw_text'].apply(f)
```

```python
expression = expression.apply(g)
```

```python
script_lines['expression'] = expression
```
