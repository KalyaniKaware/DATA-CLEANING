```python
import pandas as pd
import datetime
import re
import numpy as np
```

```python
billboard = pd.read_csv(r"C:\Users\kalya\Desktop\practise\Tidy data datasets\billboard.csv", encoding = 'unicode_escape')
billboard.head()
```

```python
billboard.columns
```

```python
billboard = billboard.melt(id_vars = ['year', 'artist.inverted', 'track', 'time', 'genre', 'date.entered',
       'date.peaked'],var_name = 'week', value_name = 'rank')
```

```python

billboard.update(pd.DataFrame({'week': np.ravel([list(map(int,re.findall(r'\d+', x))) for x in billboard['week']])}))


#matching at least 1 digit '\d+' in raw input 'r'

#ravel will return a continuous flattened array of all input lists. Since its np, input to ravel should be int, hence map function is used
```

```python
# creating a date column from date.entered and week

billboard['date.entered'] =  pd.to_datetime(billboard['date.entered'], format='%Y/%m/%d')
```

```python
billboard = (billboard.assign(date = [(billboard['date.entered'][i]+datetime.timedelta(weeks = billboard["week"][i]-1)) for i in range(len(billboard['week']))]))
```

```python
billboard.drop(['date.entered','date.peaked','week'],inplace = True, axis = 1)
```

```python
billboard = billboard.sort_values('artist.inverted',ascending = True).reset_index(drop = True)
```

```python
values,uniques = pd.factorize(billboard.track)
```

```python
billboard['id'] = values+1
billboard.reset_index(drop=True)
billboard.head()
```

```python
ranks = billboard.filter(['id','date','rank']).dropna().sort_values(['id','date']).reset_index(drop=True)
songs = billboard.filter(['id','artist.inverted','track','time']).drop_duplicates('id').reset_index(drop=True)
```
